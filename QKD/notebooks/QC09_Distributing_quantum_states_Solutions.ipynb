<h4>Measuring in incorrect basis</h4>

Create a program for 16 bits, where Asja sends first 8 bits encoded using $X$-basis and next 8 bits encoded in $Z$-basis, but Balvis measures first 8 qubits using $Z$-basis and next 8 qubits using $X$-basis. 

#### Solution

```python
# import all necessary objects and methods for quantum circuits
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from random import randrange


def print_outcomes_in_reserve(counts):
    """
    Takes a dictionary of counts and returns a new dictionary
    with bitstrings reversed (since Qiskit uses little-endian order).
    """
    reversed_counts = {}
    
    for outcome, value in counts.items():
        reverse_outcome = outcome[::-1]   # simple and clean string reversal
        reversed_counts[reverse_outcome] = value
        
    return reversed_counts


def SendState(qc1, qc2):
    """
    This function replays the X and H gates from qc1
    onto qc2 to reproduce the same state.
    
    Assumes qc1 only contains x and h gates.
    """

    for instruction, qargs, cargs in qc1.data:
        
        if instruction.name == "x":
            qubit_index = qc1.qubits.index(qargs[0])
            qc2.x(qc2.qubits[qubit_index])
            
        elif instruction.name == "h":
            qubit_index = qc1.qubits.index(qargs[0])
            qc2.h(qc2.qubits[qubit_index])
            
        elif instruction.name == "measure":
            # Skip measurements
            continue
            
        else:
            raise Exception(f"Unsupported instruction: {instruction.name}")
```

```python
qreg = QuantumRegister(16) # quantum register with 8 qubits
creg = ClassicalRegister(16) # classical register with 8 bits

# Quantum circuit for Asja state
asja = QuantumCircuit(qreg, creg, name='Asja')

send=[] #Initial bit string to send

#Creating random bit string
for i in range(16):
    bit = randrange(2)
    send.append(bit)
    
#Apply X gate if bit is equal to 1
for i, n in enumerate(send):
    if n==1:
        asja.x(qreg[i]) # apply x-gate
#        print("x-gate is applied to the qubit with index",i)

send_str = ''.join(str(e) for e in send)

for i in range(8):
    asja.h(qreg[i]) #first 8 qubits encoded in X basis

asja.draw() #To check gates applied correctly
```

Balvis measures first 8 qubits using $Z$-basis and next 8 qubits using $X$-basis. So, he applies the Hadamard operation on qubits 8 - 15. 

```python
balvis = QuantumCircuit(qreg, creg, name='Balvis') #Defining Balvis circuit
SendState(asja, balvis) #Asja sends states to Balvis

# Balvis measures first 8 qubits using Z-basis (0 - 7) and next 8 qubits using X-basis (8 - 15)
for n in range(8, 16):
    balvis.h(qreg[n])
    
balvis.draw()
```

```python
from qiskit_aer import Aer
from qiskit import transpile

balvis.h(qreg)
balvis.measure(qreg, creg)

backend = Aer.get_backend('qasm_simulator')
transpiled = transpile(balvis, backend)
job = backend.run(transpiled, shots=1)

counts = job.result().get_counts(balvis)
received = print_outcomes_in_reserve(counts)

received_str = list(received.keys())[0]

print("Asja sent:     ", send_str)
print("Balvis received:", received_str)
```
