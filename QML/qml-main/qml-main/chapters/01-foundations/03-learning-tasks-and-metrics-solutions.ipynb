This notebook contains complete solutions for the embedded tasks in notebook `1.3`.

```python
import numpy as np
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    mean_absolute_error,
    mean_squared_error,
)
```

<a id="task1-solution"></a>

## Solution to Task 1

One reasonable set of answers is:

- A) Binary fraud detection on tabular transaction data
  - Type: supervised learning, more specifically binary classification.
  - QML candidacy: weak or unclear as a first option.
  - Justification: strong classical baselines usually perform very well on tabular data, and quantum overhead may not be justified.
- B) Denoising and generating short molecular spectra
  - Type: generative learning.
  - QML candidacy: potentially stronger exploratory case.
  - Justification: the task is distribution-oriented and the data is closer to a physical domain, so a quantum representation may be worth exploring.
- C) Predicting house prices from 30 numeric features
  - Type: supervised learning, more specifically regression.
  - QML candidacy: usually weak as a first option.
  - Justification: classical regression methods are strong, cheap to train, and easier to validate carefully.
- D) Grouping unlabeled customer-behavior traces into a few usage profiles before any manual labeling
  - Type: unsupervised learning, more specifically clustering.
  - QML candidacy: unclear.
  - Justification: the task matches an unsupervised setting, but the practical benefit of a quantum model depends heavily on how the data is encoded and whether the clustering structure is actually captured well in a low-qubit representation.
- E) Training an agent to choose the next control pulse in a sequential quantum experiment based on reward feedback
  - Type: reinforcement learning.
  - QML candidacy: unclear but interesting as an exploratory direction.
  - Justification: the problem is sequential and reward-driven rather than label-driven, so it belongs to reinforcement learning; quantum approaches may be relevant in control settings, but this is outside the main scope of the present notebook family split.

## Solution to Task 2

One reasonable set of choices is:

- A) Medical screening with strong class imbalance
  - Primary metric: Recall or F1.
  - Secondary metric: ROC-AUC.
  - Justification: raw accuracy can be misleading under imbalance, so we need metrics that reflect minority-class performance and ranking quality. If missing a positive case is especially costly, Recall may be the first number to watch.
- B) Real-valued sensor calibration
  - Primary metric: MAE.
  - Secondary metric: MSE.
  - Justification: MAE is easy to interpret in the same physical units as the target, while MSE is useful when large errors should be penalized more strongly.
- C) Generating binary patterns from a target distribution
  - Primary metric: a distribution distance such as TV distance, JS divergence, or MMD.
  - Secondary metric: another distribution-aware score for robustness.
  - Justification: the goal is to match a distribution, not to predict labels one by one, so classification-style metrics would miss the main objective.
- D) Clustering unlabeled customer sessions into behavior groups, where no ground-truth labels are available during training
  - Primary metric: Silhouette score.
  - Secondary metric: Davies-Bouldin score.
  - Justification: both are internal clustering metrics that can be computed without labels. If evaluation labels later become available, an external score such as ARI could also be reported, but it is not the natural primary choice from the problem statement.
- E) Binary classifier for loan approval, where the ranking of applicants matters before fixing a final decision threshold
  - Primary metric: ROC-AUC.
  - Secondary metric: Precision or Recall, depending on operational priorities.
  - Justification: the scenario explicitly says ranking matters before threshold selection, which is exactly the type of setting where ROC-AUC is informative. Accuracy alone would hide how well the score ordering behaves.

## Solution to Task 3

Here we compare two classification thresholds built from the same score values, and two regression candidates with different error patterns.

```python
y_true_cls_new = np.array([0, 1, 1, 1, 0, 0, 1, 0])
y_score_cls_new = np.array([0.15, 0.85, 0.70, 0.48, 0.35, 0.60, 0.90, 0.20])
thresholds = [0.40, 0.70]

y_true_reg_new = np.array([0.5, 1.7, 2.9, 4.2])
y_pred_reg_A = np.array([0.8, 2.0, 3.2, 4.5])
y_pred_reg_B = np.array([0.5, 1.7, 2.9, 5.1])

auc_new = roc_auc_score(y_true_cls_new, y_score_cls_new)

print("Classification comparison")
for threshold in thresholds:
    y_pred_from_score = (y_score_cls_new >= threshold).astype(int)
    acc = accuracy_score(y_true_cls_new, y_pred_from_score)
    prec = precision_score(y_true_cls_new, y_pred_from_score)
    rec = recall_score(y_true_cls_new, y_pred_from_score)
    f1 = f1_score(y_true_cls_new, y_pred_from_score)
    print(f"threshold={threshold:.2f} -> Accuracy={acc:.3f}, Precision={prec:.3f}, Recall={rec:.3f}, F1={f1:.3f}")

print(f"ROC-AUC from the score values: {auc_new:.3f}")

mse_A = mean_squared_error(y_true_reg_new, y_pred_reg_A)
mae_A = mean_absolute_error(y_true_reg_new, y_pred_reg_A)
mse_B = mean_squared_error(y_true_reg_new, y_pred_reg_B)
mae_B = mean_absolute_error(y_true_reg_new, y_pred_reg_B)

print("\nRegression comparison")
print(f"Model A -> MSE={mse_A:.3f}, MAE={mae_A:.3f}")
print(f"Model B -> MSE={mse_B:.3f}, MAE={mae_B:.3f}")
```

A reasonable interpretation is:
- For classification, the threshold `0.40` gives higher recall because it predicts positives more aggressively, while `0.70` gives higher precision because it is more conservative. The ROC-AUC stays the same because it depends on the score ranking, not on the chosen threshold.
- For regression, **MAE** prefers model B because three predictions are exact and only one error is large, so the average absolute error stays relatively small. **MSE** prefers model A because squaring the errors penalizes the single large mistake in model B much more strongly.

## Solution to Task 4

Two standard examples are:

- **Misleading accuracy**: suppose 95% of examples belong to class 0 and only 5% to class 1. A weak model that predicts class 0 almost all the time can still achieve about 95% accuracy while failing badly on the minority class. In that case, accuracy hides the real failure mode.
- **Different preferences under MSE and MAE**: imagine Model A makes many small errors, while Model B is usually accurate but occasionally makes one very large mistake. MAE may judge them as relatively similar, but MSE will penalize the rare large error much more strongly and may clearly prefer Model A.

## Solution to Task 5

We implement Total Variation distance and compare both toy generative outputs against the target distribution.

```python
p_target = np.array([0.10, 0.20, 0.30, 0.40])
p_model_A = np.array([0.12, 0.18, 0.33, 0.37])
p_model_B = np.array([0.05, 0.25, 0.20, 0.50])

def tv_distance(p, q):
    return 0.5 * np.sum(np.abs(p - q))

tv_A = tv_distance(p_target, p_model_A)
tv_B = tv_distance(p_target, p_model_B)

print(f"TV(target, model A) = {tv_A:.3f}")
print(f"TV(target, model B) = {tv_B:.3f}")
```

Model A is closer to the target because its TV distance is smaller. In this setting, a smaller TV distance means the two probability distributions place more similar total probability mass across the same outcomes.
