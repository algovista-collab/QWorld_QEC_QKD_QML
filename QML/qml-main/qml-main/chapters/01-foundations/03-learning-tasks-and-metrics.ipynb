# Learning Tasks and Metrics

_Prepared by Claudia Zendejas-Morales_

Machine learning questions are not all of the same kind, so they should not be evaluated in the same way. Before asking whether a classical or quantum model is promising, we must first identify the **learning task**: are we classifying, predicting a real value, discovering structure without labels, or modeling a probability distribution? Once the task is clear, the next step is to choose metrics that actually reflect the goal of the problem. In this notebook, we organize common QML tasks into families, review the metrics most often used for each one, and show with small examples why metric choice can change the conclusion of an experiment.

## Task families

A useful organization for this tutorial is into **supervised**, **unsupervised**, and **generative** learning. This is a practical choice for QML because these three families already cover most of the workflows we will study later: predicting labels or values, discovering structure in unlabeled data, and learning distributions that can be sampled.

A more general machine learning view is slightly broader:
- **Supervised learning** uses labeled examples $(\vec{x}, y)$. The model learns a map from inputs to targets. Common subfamilies include **binary classification**, **multiclass classification**, and **regression**. In applications, this covers tasks such as fraud detection, medical diagnosis, sentiment classification, house-price prediction, or estimating a physical parameter from experimental data.
- **Unsupervised learning** works without target labels. The goal is to uncover useful structure in the data itself. Common examples include **clustering**, **dimensionality reduction**, **representation learning**, and **anomaly detection**. In QML, this can appear when we want compressed embeddings, latent representations, or ways to separate patterns before any label is introduced.
- **Generative learning** aims to model a target probability distribution so that the model can assign probabilities or produce new samples from that distribution. This includes tasks such as **density modeling**, **data generation**, and **distribution matching**. In the quantum setting, this is where models such as **QCBMs** and **QGANs** naturally appear.

In broader ML discussions, generative modeling is often grouped together with unsupervised learning because it usually does not rely on labels. Here, however, it is useful to separate it explicitly: the goal, the output object, and the evaluation metrics are different enough that generative models deserve their own track.

It is also important to note that **reinforcement learning** is another major family of machine learning. There, an agent interacts with an environment, receives rewards, and learns a policy for sequential decision making. Quantum variants of reinforcement learning exist, but they are outside the scope of this tutorial, so we will not treat them as one of the main families here.

In QML, the task family matters because it changes what the circuit output means, what data encoding is appropriate, and how performance should be evaluated. A model that is reasonable for classification may be a poor fit for generative modeling, even if the quantum circuit looks similar.

A good rule of thumb is this: first ask what object the model is supposed to produce, and only then decide how to score it. A classifier returns labels or class scores, a regressor returns numbers, an unsupervised model returns structure, and a generative model returns a distribution or samples.

<p align="center">
  <img style="display:block; margin-left: auto; margin-right: auto;" src="./images/metrics-by-task-map.png" alt="Map from learning tasks to typical outputs and evaluation metrics" width="100%"/>
</p>
<p align="center"><em>Different task families lead naturally to different outputs, so they also require different evaluation metrics.</em></p>

So the metric is not just a reporting detail. It expresses what counts as success in the experiment, and it should be chosen before we compare models or make claims about QML performance.

For clustering without labels, internal metrics such as <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html">Silhouette score</a> and <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.davies_bouldin_score.html">Davies-Bouldin score</a> are often more natural than accuracy-style metrics. If reference labels become available only for evaluation, then external scores such as <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.adjusted_rand_score.html">Adjusted Rand Index (ARI)</a> can also be reported.

To make these choices more concrete, the next worked example computes a few standard metrics on small classification and regression datasets. The goal is not to build a model here, but to see how different metrics summarize the same predictions in different ways.

```python
import numpy as np
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    mean_absolute_error,
    mean_squared_error,
)
```

```python
# Toy arrays for a complete metric example
y_true_cls = np.array([0, 1, 1, 0, 1, 0, 1, 1])
y_pred_cls = np.array([0, 1, 0, 0, 1, 0, 1, 0])
y_score_cls = np.array([0.10, 0.80, 0.40, 0.20, 0.75, 0.30, 0.90, 0.45])

y_true_reg = np.array([1.2, 2.0, 3.1, 4.4])
y_pred_reg = np.array([1.0, 2.4, 2.8, 4.0])
```

```python
acc = accuracy_score(y_true_cls, y_pred_cls)
f1 = f1_score(y_true_cls, y_pred_cls)
auc = roc_auc_score(y_true_cls, y_score_cls)

mse = mean_squared_error(y_true_reg, y_pred_reg)
mae = mean_absolute_error(y_true_reg, y_pred_reg)

print(f"Accuracy: {acc:.3f}")
print(f"F1 score: {f1:.3f}")
print(f"ROC-AUC: {auc:.3f}")
print(f"MSE: {mse:.3f}")
print(f"MAE: {mae:.3f}")
```

The classification example above shows three different views of the same predictions: **Accuracy** measures raw correctness, **F1** balances precision and recall, and **ROC-AUC** checks whether the score values rank positive examples above negative ones. In regression, **MSE** and **MAE** are both error averages, but they react differently to large mistakes because MSE squares the residuals. This is why one dataset can make two models look similar under MAE but quite different under MSE.

<div style="background:#fff1d6;color:#111111;border-left:6px solid #d98500;padding:12px 14px;border-radius:6px;">
<b>Accuracy</b><br/>
For a classification dataset with true labels $y_i$ and predicted labels $\hat{y}_i$, accuracy is the fraction of correct predictions:
$$
\mathrm{Accuracy} = \frac{1}{N}\sum_{i=1}^{N} \mathbf{1}(\hat{y}_i = y_i)
$$
Here, $\mathbf{1}(\hat{y}_i = y_i)$ is the indicator function: it equals $1$ when the prediction is correct and $0$ otherwise.
This metric is easy to interpret, but it can be misleading when one class is much more common than the others. In that case, a weak model can obtain high accuracy simply by predicting the majority class most of the time, even if it performs poorly on the rare but important examples.
</div>

<div style="background:#fff1d6;color:#111111;border-left:6px solid #d98500;padding:12px 14px;border-radius:6px;">
<b>Precision, Recall, and F1</b><br/>
For binary classification, let <i>TP</i>, <i>FP</i>, and <i>FN</i> denote true positives, false positives, and false negatives. Then
$$
\mathrm{Precision} = \frac{TP}{TP+FP}, \qquad \mathrm{Recall} = \frac{TP}{TP+FN}
$$
The F1 score is the harmonic mean of precision and recall:
$$
F_1 = 2\,\frac{\mathrm{Precision}\cdot\mathrm{Recall}}{\mathrm{Precision}+\mathrm{Recall}}
$$
This is useful when we want one score that rewards a good balance between catching positives and avoiding too many false alarms.
</div>

<div style="background:#fff1d6;color:#111111;border-left:6px solid #d98500;padding:12px 14px;border-radius:6px;">
<b>MAE and MSE</b><br/>
For regression targets $y_i$ and predictions $\hat{y}_i$, the mean absolute error and mean squared error are
$$
\mathrm{MAE} = \frac{1}{N}\sum_{i=1}^{N} |y_i-\hat{y}_i|
$$
$$
\mathrm{MSE} = \frac{1}{N}\sum_{i=1}^{N} (y_i-\hat{y}_i)^2
$$
Both average prediction error, but MSE penalizes large mistakes more strongly because the residuals are squared, while MAE stays closer to the typical absolute deviation.
</div>
