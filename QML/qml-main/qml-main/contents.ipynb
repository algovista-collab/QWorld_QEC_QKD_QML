<h1 align="left" style="color: #e99b0b;"> Welcome </h1>

Welcome to QWorld's Quantum Machine Learning (QML) module.

This tutorial combines conceptual foundations with practical notebook-based exercises using **Qiskit Machine Learning** and **PennyLane**.

**Prerequisites**
* Completion of QBronze (or equivalent)
* Basic linear algebra, Python, and introductory machine learning

**Software requirements**
Jupyter Notebook (or JupyterLab), Python 3, Qiskit, Qiskit Machine Learning, and PennyLane.

Setup instructions are provided in Section 0 of the Table of Contents.
